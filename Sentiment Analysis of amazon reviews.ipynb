# Sentiment Analysis using VaderSentiment

```python
!pip install vaderSentiment # VADER (Valence Aware Dictionary and sEntiment Reasoner)
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
se = SentimentIntensityAnalyzer()
```

```python
se.polarity_scores(corpus.iloc[9997]['review'])
```

```python
s = 'The movie was worst movies i have ever seen'
```

```python
se.polarity_scores(s)
```

```python
corpus['Polarity Scores'] = corpus['review'].apply(lambda x: se.polarity_scores(x))
```

```python
corpus
```

```python
corpus['compound'] = corpus['Polarity Scores'].apply(lambda x: x['compound'])
```

```python
corpus
```

```python
corpus['Sentiment']=corpus['compound'].apply(lambda x: 'pos' if x>=0 else 'neg')
```

```python
corpus
```

```python
corpus.iloc[9997]
```

```python
corpus.iloc[9997]['review']
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
y_true=corpus['label']
y_pred=corpus['Sentiment']
```

```python
confusion_matrix(y_true, y_pred)
```

```python
print(classification_report(y_true, y_pred))
```

```python
print(accuracy_score(y_true, y_pred))
```

Accuracy of Sentiment Analysis using vader sentiment is 71.29%

# Sentiment Analysis using Hugging face


```python
reviews = corpus['review'].tolist()
```

```python
reviews
```

```python
from transformers import pipeline
```

```python
# Load the sentiment analysis pipeline with a small efficient model
sentiment_pipeline = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

```python
# Perform sentiment analysis
results = sentiment_pipeline(reviews)
```

```python
results
```

```python
# Print results
for review, result in zip(reviews, results):
    print(f"Review: {review}\nSentiment: {result['label']}, (Confidence: {result['score']:.2f})\n")
```

```python
corpus['SentimentHF'] = [result['label'] for result in results]
```

```python
corpus.head()
```

```python
corpus.iloc[:,-2:]
```

```python
# Compare the two columns
corpus['Match'] = corpus['label'].map({'pos': 'POSITIVE', 'neg': 'NEGATIVE'}) == corpus['SentimentHF']
```

```python
corpus['Match'].value_counts()
```

```python
# Calculate the percentage of matches
match_percentage = corpus['Match'].mean() * 100
print(f"Percentage of matches: {match_percentage:.2f}%")
```

Accuracy of Sentiment Analysis using hugging face is 90.13%
